```python
# Dependencies
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import sem

# Set the seed so our data is reproducible
random.seed(42)
```

```python
# Sample versus population example fuel economy
fuel_economy = pd.read_csv('../Resources/2019_fuel_economy.csv')

# First overview the data set - how many factors, etc.
print(fuel_economy.head())
```

```python
# Calculate the summary statistics and plot the histogram of the entire population data
mean = round(fuel_economy['Combined_MPG'].mean(), 2)
st_dev = round(fuel_economy['Combined_MPG'].std(),2)

print(f"The mean MPG of all vehicles is: {mean}")
print(f"The standard deviation of all vehicle's MPG is: {st_dev}")
plt.hist(fuel_economy['Combined_MPG'], bins=15)
plt.xlabel("Fuel Economy (MPG)")
plt.ylabel("Number of Vehicles")
plt.show()
```
